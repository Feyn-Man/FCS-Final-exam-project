```python
import pandas as pd
import re
import numpy as np
```

```python
#upload of datasets and first inspection

movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")

movies
```

```python
credits.head()
```

### 1. For each movie compute the number of cast members

```python
cast_no = []

for i in range(len(credits)):
    cast = credits.cast[i]
    cast = eval(cast)
    cast_no.append(len(cast))
    
credits['cast no'] = cast_no
credits.head()
```

### 2. How many movies do not have a homepage?

```python
homepage = movies.homepage
tot = len(homepage)
no_hp = tot - len(homepage.dropna())
no_hp
```

### 4. Extract the domain of each homepage.

```python
def extract_domain(address):
    pattern = "([\w]+)(\.([\w]+))+" #regular expression to find the domain
    return re.search(pattern, address).group()

url = list(homepage.dropna())

for string in url:
    domain = extract_domain(string)
    print(domain)
```

### 5.Extract a set of normalized tables.

```python
genres = pd.DataFrame(eval(movies.genres[0]))
genres['id_movie'] = movies.id[0]

#genres1 = pd.DataFrame(eval(movies.genres[3971]))
#if genres1.empty:
    #genres1 = genres1.append({'id': np.nan, 'name': np.nan, 'id_movie': movies.id[3971]}, ignore_index = True)
#    genres1 = genres1.append({'id_movie': movies.id[3971]}, ignore_index = True)
    

#genres1.id_movie = genres1.id_movie.astype(np.int64)
#print(type(movies.id[0]))
#print(type(genres1.id_movie[0]))
#movies.loc[[3971]]
#genres1
#pd.concat([genres, genres1], ignore_index = True, join_axes = [genres.columns])

for i in range(len(movies)):
    
    if i == max(range(len(movies))):
        break
    
    genres_i = pd.DataFrame(eval(movies.genres[i+1]))

    if genres_i.empty:    #handling of empty dataframes
        genres_i = genres_i.append({'id_movie': movies.id[i+1]}, ignore_index = True)
        genres_i.id_movie = genres_i.id_movie.astype(np.int64)

    else:
        genres_i['id_movie'] = movies.id[i+1]
        
        
    genres = pd.concat([genres, genres_i], ignore_index = True, join_axes = [genres.columns])


genres
```

```python
genres.columns = ['id_genre', 'name_genre', 'id']
genres
```

```python
keywords = pd.DataFrame(eval(movies.keywords[0]))
keywords['id_movie'] = movies.id[0]


for i in range(len(movies)):
    
    if i == max(range(len(movies))):
        break
    
    keywords_i = pd.DataFrame(eval(movies.keywords[i+1]))

    if keywords_i.empty:
        keywords_i = keywords_i.append({'id_movie': movies.id[i+1]}, ignore_index = True)
        keywords_i.id_movie = keywords_i.id_movie.astype(np.int64)

    else:
        keywords_i['id_movie'] = movies.id[i+1]
     
    #print(keywords_i)
    keywords = pd.concat([keywords, keywords_i], ignore_index = True, join_axes = [keywords.columns])


keywords
```

```python
keywords.columns = ['id_kw', 'name_kw', 'id']
keywords
```

```python
prod_comp = pd.DataFrame(eval(movies.production_companies[0]))
prod_comp['id_movie'] = movies.id[0]


for i in range(len(movies)):
    
    if i == max(range(len(movies))):
        break
    
    prod_comp_i = pd.DataFrame(eval(movies.production_companies[i+1]))

    if prod_comp_i.empty:
        prod_comp_i = prod_comp_i.append({'id_movie': movies.id[i+1]}, ignore_index = True)
        prod_comp_i.id_movie = prod_comp_i.id_movie.astype(np.int64) 

    else:
        prod_comp_i['id_movie'] = movies.id[i+1]
        
    prod_comp = pd.concat([prod_comp, prod_comp_i], ignore_index = True, join_axes = [prod_comp.columns])


prod_comp
```

```python
prod_comp.columns = ['id_company', 'company_name', 'id']
prod_comp
```

```python
prod_countries = pd.DataFrame(eval(movies.production_countries[0]))
prod_countries['id_movie'] = movies.id[0]


for i in range(len(movies)):
    
    if i == max(range(len(movies))):
        break
    
    prod_countries_i = pd.DataFrame(eval(movies.production_countries[i+1]))

    if prod_comp_i.empty:
        prod_countries_i = prod_countries_i.append({'id_movie': movies.id[i+1]}, ignore_index = True)
        prod_countries_i.id_movie = prod_countries_i.id_movie.astype(np.int64)

    else:
        prod_countries_i['id_movie'] = movies.id[i+1]
        
    prod_countries = pd.concat([prod_countries, prod_countries_i], ignore_index = True, join_axes = [prod_countries.columns])


prod_countries
```

```python
prod_countries.columns = ['iso_3166_1', 'country', 'id']
prod_countries
```

```python
languages = pd.DataFrame(eval(movies.spoken_languages[0]))
languages['id_movie'] = movies.id[0]


for i in range(len(movies)):
    
    if i == max(range(len(movies))):
        break
    
    languages_i = pd.DataFrame(eval(movies.spoken_languages[i+1]))

    if languages_i.empty:
        languages_i = languages_i.append({'id_movie': movies.id[i+1]}, ignore_index = True)
        languages_i.id_movie = languages_i.id_movie.astype(np.int64)

    else:
        languages_i['id_movie'] = movies.id[i+1]
        
    languages = pd.concat([languages, languages_i], ignore_index = True, join_axes = [languages.columns])


languages
```

In the following the same procedure is applied to the credits dataframe.

```python
cast = pd.DataFrame(eval(credits.cast[0]))
cast['movie_id'] = credits.movie_id[0]


for i in range(len(credits)):
    
    if i == max(range(len(credits))):
        break
    
    cast_i = pd.DataFrame(eval(credits.cast[i+1]))

    if cast_i.empty:
        cast_i = cast_i.append({'movie_id': credits.movie_id[i+1]}, ignore_index = True)
        cast_i.movie_id = cast_i.movie_id.astype(np.int64)

    else:
        cast_i['movie_id'] = credits.movie_id[i+1]
        
    cast = pd.concat([cast, cast_i], ignore_index = True, join_axes = [cast.columns])


cast
```

```python
crew = pd.DataFrame(eval(credits.crew[0]))
crew['movie_id'] = credits.movie_id[0]


for i in range(len(credits)):
    
    if i == max(range(len(credits))):
        break
    
    crew_i = pd.DataFrame(eval(credits.crew[i+1]))

    if crew_i.empty:
        crew_i = crew_i.append({'movie_id': credits.movie_id[i+1]}, ignore_index = True)
        crew_i.movie_id = crew_i.movie_id.astype(np.int64)

    else:
        crew_i['movie_id'] = credits.movie_id[i+1]
        
    crew = pd.concat([crew, crew_i], ignore_index = True, join_axes = [crew.columns])


crew
```

### 10. For each company and each decade, compute the overall revenue

```python
movies['decade'] = ( movies.year // 10 )*10
pd.merge(movies, prod_comp).dropna(subset =['company_name']).groupby(['company_name', 'decade']).sum()['revenue']
```

### 11. For each decade, compute the company with maximum revenue

```python
pd.merge(movies, prod_comp).dropna(subset =['company_name']).groupby(['decade'])['revenue', 'company_name'].max()
```

### 12. In each year, how many movies have revenue smaller than the budget?

```python
movies[movies.gross_margin < 0].groupby('year').size()
```
