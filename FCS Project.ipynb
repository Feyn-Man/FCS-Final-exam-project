### 1. For each movie compute the number of cast members

```python
cast_no = [] #this list will contain the number of cast members and will be attached to the movies df

for i in range(len(credits)):
    cast_no.append(len(eval(credits.cast[i])))  #eval function transforms string-list into a list
    
credits['cast no'] = cast_no
credits.head()
```

### 3. For each year, how many movies do not have a homepage?

```python
movies['year'] = pd.to_datetime(movies['release_date'].dropna()).dt.year  #year extraction

movies[movies['homepage'].isnull()].groupby('year').size()
```

### 4. Extract the domain of each homepage.

```python
def extract_domain(address):
    pattern = "([\w\-]+)(\.([\w\-]+))+" #regular expression to find the domain
    return re.search(pattern, address).group()

url = list(movies.homepage.dropna())

for string in url:
    domain = extract_domain(string)
    print(domain)
```

### 5.Extract a set of normalized tables.

In this point the columns of dataframes stored as lists of dictionaries are extracted and stored in new dataframes in a normal form.

```python
genres = pd.DataFrame(eval(movies.genres[0])) #initializing of the genres dataframe
genres['id_movie'] = movies.id[0]

for i in range(len(movies)):    #in this cycle is created the total genres df for each movie
    
    if i == max(range(len(movies))):
        break
    
    genres_i = pd.DataFrame(eval(movies.genres[i+1]))

    if genres_i.empty:    #handling of empty dataframes (in case genres_i is empty we can't add column like in the else statement)
        genres_i = genres_i.append({'id_movie': movies.id[i+1]}, ignore_index = True)
        genres_i.id_movie = genres_i.id_movie.astype(np.int64)

    else:
        genres_i['id_movie'] = movies.id[i+1]
        
        
    genres = pd.concat([genres, genres_i], ignore_index = True, join_axes = [genres.columns])    #union of the new df with the previous


genres
```

```python
genres.columns = ['id_genre', 'name_genre', 'id']
genres
```

```python
prod_comp = pd.DataFrame(eval(movies.production_companies[0]))
prod_comp['id_movie'] = movies.id[0]


for i in range(len(movies)):
    
    if i == max(range(len(movies))):
        break
    
    prod_comp_i = pd.DataFrame(eval(movies.production_companies[i+1]))

    if prod_comp_i.empty:
        prod_comp_i = prod_comp_i.append({'id_movie': movies.id[i+1]}, ignore_index = True)
        prod_comp_i.id_movie = prod_comp_i.id_movie.astype(np.int64) 

    else:
        prod_comp_i['id_movie'] = movies.id[i+1]
        
    prod_comp = pd.concat([prod_comp, prod_comp_i], ignore_index = True, join_axes = [prod_comp.columns])


prod_comp
```
